```python

%matplotlib inline

import matplotlib.pyplot as plt
import random
from collections import Counter
import math
import numpy

EleitoresTotal = []
n = 200
lista_Pchapeu = []
soma = 0

for ciclo in range (2000):
    amostra = []
    ContagemA_amostra = 0
    ContagemB_amostra = 0
    for i in range (3000): #eleitores de A
        EleitoresTotal.append(0)
    for j in range (7000): #ELeitores de B
        EleitoresTotal.append(1)

    for k in range (n): #amostra com reposição
        aleatorio = random.randint(0,9999) #selecionando um indice aleatório
        amostra.append(EleitoresTotal[aleatorio]) #adicionando o eleitor referente ao índice sorteado.

        # Calculando a proporção amostral de votos no candidato A
        if EleitoresTotal[aleatorio] == 0:
            ContagemA_amostra += 1
        else:
            ContagemB_amostra += 1

    pchapeu = ContagemA_amostra / n
    soma = soma + pchapeu
    lista_Pchapeu.append(pchapeu)


media = soma / 2000
print('')
print('Para uma amostra de n = 200, obtem-se como resultado:')
print('Média =', media)

Armazenamento = lista_Pchapeu
DesvioAcumulado = 0
#Cálculo desvio padrão
for desvio in range (2000):
  diferenca = lista_Pchapeu[desvio] - media
  square = diferenca ** 2
  raiz = square**(1/2)
  DesvioAcumulado = DesvioAcumulado + raiz
  DesvioPadrao = DesvioAcumulado / 2000
  
print('Desvio Padrão =', DesvioPadrao)
  
 
  



#Plotar gráficos:

#Gráfico para n = 200
print('A representação gráfica do “p-chapéu” para cada ciclo de simulação encontra-se a seguir:')
grap = plt.figure(1, figsize=(12,7))

plotgrafico = grap.add_subplot(211)
plotgrafico.plot(range(2000), lista_Pchapeu, color='r', label='n = 200')
plt.legend()
plt.title('"Pchapéu" em cada simulação')
plt.ylabel('Probabilidade de "Pchapéu"')

plotgrafico = grap.add_subplot(212)
plt.hist(Armazenamento, bins = 40, alpha = 1, color = 'r', label = 'n = 200')
plt.legend()
plt.title('Histograma para n = 200')
plt.xlabel('Probabilidade de "pchapéu"')
plt.ylabel('Número de ocorrências')
plt.show()



EleitoresTotal = []
n = 2000
lista_Pchapeu = []
soma = 0

for ciclo in range (2000):
    amostra = []
    ContagemA_amostra = 0
    ContagemB_amostra = 0
    for i in range (3000): #eleitores de A
        EleitoresTotal.append(0)
    for j in range (7000): #ELeitores de B
        EleitoresTotal.append(1)

    for k in range (n): #amostra com reposição
        aleatorio = random.randint(0,9999) #selecionando um indice aleatório
        amostra.append(EleitoresTotal[aleatorio]) #adicionando o eleitor referente ao índice sorteado.

        # Calculando a proporção amostral de votos no candidato A
        if EleitoresTotal[aleatorio] == 0:
            ContagemA_amostra += 1
        else:
            ContagemB_amostra += 1

    pchapeu = ContagemA_amostra / n
    soma = soma + pchapeu
    lista_Pchapeu.append(pchapeu)


media = soma / 2000
print('Para uma amostra de n = 2000, obtem-se como resultado:')
print('Média =', media)

DesvioAcumulado = 0
#Cálculo desvio padrão
for desvio in range (2000):
  diferenca = lista_Pchapeu[desvio] - media
  square = diferenca ** 2
  raiz = square**(1/2)
  DesvioAcumulado = DesvioAcumulado + raiz
  DesvioPadrao = DesvioAcumulado / 2000
  
print('Desvio Padrão =', DesvioPadrao)
  


#gráfico para n = 2000
print('A representação gráfica do “p-chapéu” para cada ciclo de simulação encontra-se a seguir:')
grap = plt.figure(2, figsize=(12,7))

plotgrafico = grap.add_subplot(211)
plotgrafico.plot(range(2000), lista_Pchapeu,color='b', label='n = 2000')
plt.legend()
plt.title('"Pchapéu" em cada simulação')
plt.ylabel('Probabilidade de "Pchapéu"')


plotgrafico = grap.add_subplot(212)
plt.hist(lista_Pchapeu, bins = 75, color ='b', label = 'n = 2000')
plt.legend()
plt.title('Histograma para n = 2000')
plt.xlabel('Probabilidade de "pchapéu"')
plt.ylabel('Número de ocorrências')


leg = plotgrafico.legend()

plt.show()

grap = plt.figure(2, figsize=(12,7))
plt.hist(Armazenamento, bins = 80, alpha = 0.5, color = 'r', label = 'n = 200')
plt.hist(lista_Pchapeu, bins = 35, alpha = 0.8, color ='b', label = 'n = 2000')
plt.legend()
plt.title('Histograma comparativo entre as simulações')
plt.xlabel('Percentual de "pchapéu"')
plt.ylabel('Número de ocorrências')
plt.show()

print("""No código desenvolvido utilizamos duas amostras cujos tamanhos são de 200 e 2000, 
além de serem realizadas 2000 simulações para cada, obtendo uma lista de 'pchapeu' a ser
analisada.
Observa-se que o primeiro caso apresenta maior dispersão dos resultados, o que reflete no 
maior desvio padrão, isso devido ao menor número de amostra coletados. Enquanto, o segundo 
por possuir um maior número de amostras resulta em uma menor dispersão e ,consequentemente, 
menor desvio. Apesar disso, ambos os casos resultam em um valor de média muito próximo a 
proporção populacional, fator que demonstra que apesar da diferença de tamanho das amostra 
ambas convergem a um valor comum, no caso a proporção populacional a qual se quer estimar.
Então, os dois resultados diferenciam-se quanto a precisão de suas informações.""")
```
